```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from payoffs import forward_payoff

S = np.linspace(0, 200, 100)
K = 100
plt.plot(S, forward_payoff(S, K))
```

```python
from payoffs import call_payoff

S = np.linspace(0, 200, 200)
K = 100

plt.figure(figsize=(6,4))
plt.plot(S, call_payoff(S, K))
plt.axvline(K, linestyle="--", alpha=0.5)
plt.title("European Call Payoff")
plt.xlabel("Underlying price $S_T$")
plt.ylabel("Payoff")
plt.grid(True)
plt.show()
```

```python
from payoffs import put_payoff

S = np.linspace(0, 200, 200)
K = 100

plt.figure(figsize=(6,4))
plt.plot(S, put_payoff(S, K))
plt.axvline(K, linestyle="--", alpha=0.5)
plt.title("European Put Payoff")
plt.xlabel("Underlying price $S_T$")
plt.ylabel("Payoff")
plt.grid(True)
plt.show()
```
